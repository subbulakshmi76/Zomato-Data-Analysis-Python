```python
import pandas as pd

data = {
    "Restaurant": ["A2B", "KFC", "Dominos", "Pizza Hut", "Subway"],
    "City": ["Chennai", "Coimbatore", "Chennai", "Madurai", "Trichy"],
    "Rating": [4.2, 4.0, 3.8, 4.1, 3.9],
    "Cost": [500, 700, 600, 800, 400]
}

df = pd.DataFrame(data)

df
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

df.plot(x="Restaurant", y="Rating", kind="bar")

plt.title("Restaurant Ratings")
plt.show()
```

```python
print("Average Rating:", df["Rating"].mean())
```

```python
df.sort_values(by="Rating", ascending=False)
```
